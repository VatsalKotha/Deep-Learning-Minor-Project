```python
from keras.models import Sequential
from keras.layers import Dense

import pandas as pd

ds = pd.read_csv("/content/diabetes.csv")
print(ds.shape)

X = ds.iloc[:,0:8]
Y = ds.iloc[:,8]

print(X.shape)
print(Y.shape)
```

```python
# first model
model = Sequential()
# adding 3 dense layers
model.add(Dense(12, input_dim=8, kernel_initializer = 'uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

# compiliing model

model.compile(loss = 'binary_crossentropy', optimizer='adam',metrics=['accuracy'])
#fit the model
model.fit(X, Y, epochs = 20, batch_size=10, verbose = 2)
```

```python
#evaluating the model
scores = model.evaluate(X, Y)
print('%s: %.2f%%' % (model.metrics_names[1], scores[1]*100))
```

```python
model.summary()
```

```python
# calculating predictions
predictions = model.predict(X)
#round predictions
rounded = [round(x[0]) for x in predictions]
print(predictions)
print(rounded)
```

```python
# training and evaluating a simple MLP on the Reuters newswire topic classification task

from __future__ import print_function
import numpy as np
np.random.seed(1337)

from keras.datasets import reuters
from keras.layers import Dropout, Activation
from keras.utils import to_categorical
from keras.preprocessing.text import Tokenizer

#define parameters
max_words = 1000
batch_size = 32
epochs = 5


#load data
(X_train, y_train), (X_test, y_test) = reuters.load_data(num_words=max_words, test_split=0.2)
print(len(X_train))
print(len(X_test))

nb_classes = np.max(y_train)+1
print('Vectorizing the data')
tokenizer = Tokenizer(num_words=max_words)
X_train = tokenizer.sequences_to_matrix(X_train, mode = 'binary')
X_test = tokenizer.sequences_to_matrix(X_test, mode='binary')
print(X_train.shape)
print(X_test.shape)

Y_train = to_categorical(y_train, nb_classes)
Y_test = to_categorical(y_test, nb_classes)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model1 = Sequential()

model1.add(Dense(512, input_shape =(max_words, )))
model1.add(Activation('relu'))
model1.add(Dropout(0.5))
model1.add(Dense(nb_classes))
model1.add(Activation('softmax'))

model1.compile(loss='categorical_crossentropy', optimizer='adam')
history = model1.fit(X_train, Y_train, epochs=epochs, batch_size = batch_size, verbose = 1, validation_split = 0.1)
score = model1.evaluate(X_test, Y_test, batch_size=batch_size, verbose=1)
print(score)
print(score)
```
